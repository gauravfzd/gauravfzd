# Outlier Analysis

```python
# Interquartile Range (IQR) Method,
#numerical columns
num_cols = ["views", "likes", "dislikes", "comment_count"]

outlier_summary = {}
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    outlier_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
    outlier_summary[col] = {"Lower Bound": lower_bound, "Upper Bound": upper_bound, "Outliers": outlier_count}

outlier_summary
```

```python
import numpy as np
# Compute median views for realistic replacement
median_views = df["views"].median()

# Apply capping: replace extreme high values with the upper bound
#df["views"] = np.where(df["views"] > 1661093, 1661093, df["views"])

# Ensure no negative values (replace negatives with median)
df["views"] = np.where(df["views"] < 0, median_views, df["views"])

# Verify the changes
df["views"].describe()
```

```python
#Do certain categories have a higher chance of going viral?
```

```python
df.groupby("category_id")['is_viral'].sum()
```

```python
df["category_id"].value_counts()
```

```python
category_viral_count = df.groupby("category_id")["is_viral"].sum()
category_total_count = df["category_id"].value_counts()
category_viral_percentage = (category_viral_count / category_total_count) * 100
```

```python
category_viral_percentage
```

```python
category_viral_percentage = category_viral_percentage.sort_values(ascending=False)
category_viral_percentage
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the percentage of viral videos by category
plt.figure(figsize=(12,6))
sns.barplot(x=category_viral_percentage.index, y=category_viral_percentage.values, palette="coolwarm")
plt.xlabel("Category ID")
plt.ylabel("Percentage of Viral Videos (%)")
plt.title("Percentage of Viral Videos by Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# Compute correlation between views, likes, and dislikes
correlation_matrix = df[["views", "likes", "dislikes"]].corr()
correlation_matrix
```

```python
# Check unique category IDs in the dataset
df["category_id"].unique()
# Mapping of category_id to category names (based on YouTube API)
category_mapping = {
    1: "Film & Animation",
    2: "Autos & Vehicles",
    10: "Music",
    15: "Pets & Animals",
    17: "Sports",
    19: "Travel & Events",
    20: "Gaming",
    22: "People & Blogs",
    23: "Comedy",
    24: "Entertainment",
    25: "News & Politics",
    26: "Howto & Style",
    27: "Education",
    28: "Science & Technology",
    29: "Nonprofits & Activism",
    30: "Movies",
    43: "Shows"
}

# Replace category_id with actual names
df["category"] = df["category_id"].map(category_mapping)

# Verify the mapping
df[["category_id", "category"]].drop_duplicates()


# Compute engagement metrics per category
engagement_metrics = df.groupby("category")[["likes", "comment_count"]].mean()

# Add total engagement column (likes + comments)
engagement_metrics["total_engagement"] = engagement_metrics["likes"] + engagement_metrics["comment_count"]

# Sort by total engagement (highest first)
engagement_metrics = engagement_metrics.sort_values(by="total_engagement", ascending=False)

# Display top categories by engagement
engagement_metrics
```

```python
# Convert 'publish_time' to datetime format (removing 'Z' at the end if present)
df["publish_time"] = pd.to_datetime(df["publish_time"], errors="coerce")

# Extract hour and day of the week
df["publish_hour"] = df["publish_time"].dt.hour
df["publish_day"] = df["publish_time"].dt.day_name()

# Check the first few rows to verify extraction
df[["publish_time", "publish_hour", "publish_day"]].head()

# Group by publish hour and calculate average views
hourly_views = df.groupby("publish_hour")["views"].mean()

# Plot the results
plt.figure(figsize=(12,6))
sns.lineplot(x=hourly_views.index, y=hourly_views.values, marker="o", color="blue")
plt.xlabel("Hour of the Day (24-Hour Format)")
plt.ylabel("Average Views")
plt.title("Average Views by Publishing Hour")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```
